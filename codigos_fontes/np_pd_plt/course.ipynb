```python
#Sem usar pandas
def calc_rat_stats(data, industry=None):
  ratings = []
  for row in data:
    if row[3] != 'NULL' and (not industry or row[1] == industry):
      ratings.append(float(row[3]))
  max_rating = max(ratings)
  min_rating = min(ratings)
  avg_rating = sum(ratings) / len(ratings)
  return max_rating, min_rating, avg_rating

with open("movies.csv") as f:
  data = list(csv.reader(f))
  header = data[0]
  data = data[1:]

  max_rating, min_rating, avg_rating = calc_rat_stats(data)
  print(f"All records: Min rating = {min_rating}, Max rating = {max_rating}, Avg rating = {avg_rating}")

  max_rating, min_rating, avg_rating = calc_rat_stats(data, industry='Bollywood')
  print(f"Bollywood: Min rating = {min_rating}, Max rating = {max_rating}, Avg rating = {avg_rating}")

  max_rating, min_rating, avg_rating = calc_rat_stats(data, industry='Hollywood')
  print(f"Hollywood: Min rating = {min_rating}, Max rating = {max_rating}, Avg rating = {avg_rating}")
```

```python
def group(df, idx, col):
  if df[col].loc[idx] <= 3.9:
    return 'Poor'
  elif df[col].loc[idx] <= 7.9:
    return 'Average'
  else:
    return 'Good'

#solution
def grouper2(df, idx, col):
    '''
        This function returns category based on imdb_rating
    '''
    if 1 <= df[col].loc[idx] <= 3.9:
        return 'Poor'
    elif 4 <= df[col].loc[idx] <= 7.9:
        return 'Average'
    elif 8 <= df[col].loc[idx] <= 10:
        return 'Good'
    else:
        return "others"
```

```python
g = movies_df.groupby(lambda x: grouper2(movies_df, x, 'imdb_rating'))
print(g.size())

for k, v in g:
  print(k)
  print(v)
```
